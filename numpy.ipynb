```python
# compute x and y coordinates for points on a sine curve

x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
plt.title("Sine Wave Example")
plt.plot(x,y)
```

```python
#Bar Plot

x = [2,8,10]
y =[11,6,9]

x2= [3,9,11]
y2=[6,15,7]
plt.bar(x,y)
plt.bar(x2,y2, color = 'g')
plt.title('Bar Graph')
plt.xlabel('x axis ')
plt.ylabel('y axis ')
```
